```python
import requests
from bs4 import BeautifulSoup
```

```python
counter = 1
info_list = []
h1 = []
p = []
while counter<=1398:
  url = f'https://bidzer.ge/?page={counter}'
  counter+=1

  page = requests.get(url)

  try:
    page.raise_for_status()
    # print('succesfuly')
  except requests.exceptions.HTTPError as err:
    print(f"HTTP error occurred: {err}")

  soup = BeautifulSoup(page.text, 'html.parser')

  words = soup.find_all('article', class_ ='bidzer-post')


  for word in words:
    info_list.append(word.find_all('p', class_ = 'post-definition'))
    h1.append(word.select('h1 a'))
    p.append(word.find_all('p', class_ = 'post-use'))
```

```python
k = []
b = []
l = []
for i in h1:
  k.append(i[0].text.strip())
for t in info_list:
  b.append(t[0].text.strip())
for m in p:
  l.append(m[0].text.strip())
```

```python
import pandas as pd

dic = [
    'word',
    'explanation',
    'use_case'
]

df = pd.DataFrame(columns = dic)
df['word'] = k
df['explanation'] = b
df['use_case'] = l

df.head()
```

```python
from google.colab import files

df.to_csv('bizeri-info.csv', index=False)
files.download('bizeri-info.csv')
```
