```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# get url
url = 'https://en.wikipedia.org/wiki/Georgia_(country)'
```

```python
# request url from webpage

page = requests.get(url)
# check out if request was successfully
try:
  page.raise_for_status()
  print('succesfull')
except requests.exceptions.HTTPError as err:
    print(f"HTTP error occurred: {err}")
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# getting specipic table from page

table = soup.find_all('table')[2]
```

```python
# creating table titles list
table_titles = [titles.text.strip() for titles in table.find_all('th')]
```

```python
column_data = table.find_all('tr')
```

```python
df = pd.DataFrame(columns = table_titles)
```

```python
# dropping dataframe rows to avoid data duplicating
df.drop(df.index, inplace=True)

# to extract individual rows
for row in column_data[1:]:

  # getting individual row data
  row_data = [cell.text.strip() for cell in row.find_all(['td'])]

  # inserting row data in dataframe
  length = len(df)
  df.loc[length] = row_data
```
