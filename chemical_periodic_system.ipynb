```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Periodic_table'
```

```python
response = requests.get(url)
```

```python
page = bs(response.text, 'html.parser')
```

```python
spans = page.find_all('span', class_ = 'nounderlines')
```

```python
elements = []
periodic_number = []
symbol = []
weight = []
for i in range(len(spans)):
  elem = spans[i].select('a>span>span')[0].text
  periodic_number.append(spans[i].select('a>span>span')[1].text)
  symbol.append(spans[i].select('a>span>span')[2].text)
  weight.append(spans[i].select('a>span>span')[3].text)
  if '\xad' in elem:
    elem = elem.replace('\xad','')
    elements.append(elem)
  else:
    elements.append(elem)
```

```python
df = pd.DataFrame()
```

```python
len(periodic_number)
```

```python
df['elements'] = elements
df['periodic_number'] = periodic_number
df['symbols'] = symbol
df['weight'] = weight
```
